**Segunda evaliación de Física Computacional 1**

```python
# Bibliotecas a usar

import numpy as np
import matplotlib.pyplot as plt
 
from scipy.integrate import odeint
import scipy.linalg as la
 
from scipy.integrate import solve_ivp, odeint
```

**Problema 1**

```python
# Definimos la función F(t,y)

def oscilador(y, t, C, m, w, F):
        x, v = y
        dydt = [v, -2 * C * w * v - w**2 * x + (1/m) * F * np.sin(w * t)]
        return dydt

# Definimos el tiempo de integración
t = np.linspace(0, 4, 3000)

# Definimos las condiciones iniciales
y0 = [-1.0, 0.0]
F = 0.0
m = 1.0
k = 9
w = np.sqrt(k/m)
```

```python
# Caso 1 
C=0.0
y = odeint(oscilador, y0, t, args=(C,m,w,F))
# Caso 2 
C=0.1
y2 = odeint(oscilador, y0, t, args=(C,m,w,F))
# Caso 3 
C=0.2
y3 = odeint(oscilador, y0, t, args=(C,m,w,F))
# Caso 4 
C=0.4
y4 = odeint(oscilador, y0, t, args=(C,m,w,F))
# Caso 5 
C=0.7
y5 = odeint(oscilador, y0, t, args=(C,m,w,F))
# Caso 6 
C=1
y6 = odeint(oscilador, y0, t, args=(C,m,w,F))
# Caso 7 
C=2
y7 = odeint(oscilador, y0, t, args=(C,m,w,F))

# Graficamos

plt.figure(figsize=(10,5))
plt.plot(t, y[:, 0] , 'black', label='C=0.0')
plt.plot(t, y2[:, 0], 'g', label='C=0.1')
plt.plot(t, y3[:, 0], 'gray', label='C=0.2')
plt.plot(t, y4[:, 0], 'r', label='C=0.4')
plt.plot(t, y5[:, 0], 'orange', label='C=0.7')
plt.plot(t, y6[:, 0], 'purple', label='C=1.0')
plt.plot(t, y7[:, 0], 'b', label='C=2.0')
plt.legend(loc='best')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid()
plt.show()
```
